```python
# Question 1: How many people have heart disease and how many don't?
# Getting the values
heart_disease_counts = df['target'].value_counts()
print(heart_disease_counts)
```

```python
# Plotting a pie chart
heart_disease_counts.plot(kind='pie', figsize=(8, 6), autopct='%1.1f%%')
plt.title("Heart Disease Distribution")
plt.legend(["1 = Disease", "0 = No Disease"])
plt.show()
```

```python
# Question 2: People of which sex have the most heart disease?
# In the dataset, '0' represents female and '1' represents male
sex_counts = df['sex'].value_counts()
print(sex_counts)
```

```python
# Plotting a pie chart for male-female ratio
sex_counts.plot(kind='pie', figsize=(8, 6), autopct='%1.1f%%')
plt.title("Male-Female Ratio")
plt.legend(["Male", "Female"])
plt.show()
```

```python
#'0' represents 'no disease' and '1' represents 'disease'
pd.crosstab(heart_disease_counts , sex_counts)
```

```python
sns.countplot(x = 'target' , data = df , hue = 'sex')
plt.title('Heart Disease Frequency for Sex', fontsize=10);
plt.xlabel('0 = No Heart Disease , 1 = Heart Disease')
plt.show()
```

```python
#Question 5- How does blood pressure vary with age?
# Scatter plot of blood pressure (trestbps) vs age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='trestbps', data=df, color='purple')
plt.title("Blood Pressure (trestbps) vs Age", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Resting Blood Pressure (trestbps)", fontsize=12)
plt.show()
```

```python
#Question 6. Which age group has the highest cholesterol level?
# Create age groups (bins)
bins = [0, 30, 40, 50, 60, 70, 80]
labels = ['<30', '30-39', '40-49', '50-59', '60-69', '70-79']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right= False)

# Group by age group and calculate average cholesterol
age_group_cholesterol = df.groupby('age_group')['chol'].mean().reset_index()

# Plot the results
plt.figure(figsize=(10, 6))
sns.barplot(x='age_group', y='chol', data=age_group_cholesterol, palette='Blues_d')
plt.title("Average Cholesterol Level by Age Group", fontsize=16)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Average Cholesterol Level", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
#Age group between 60-69 has the highest cholestrol level
```

```python
#Question 7. What is the distribution of max heart rate among patients?
# Plot max heart rate (thalach) distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['thalach'], kde=True, bins=20, color='green')
plt.title("Distribution of Maximum Heart Rate (thalach)", fontsize=16)
plt.xlabel("Maximum Heart Rate (thalach)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
#Question 8. How many patients fall into each risk category based on cholesterol levels?
# Define cholesterol risk categories
bins = [0, 200, 240, 600]
labels = ['Desirable', 'Borderline High', 'High']
df['cholesterol_category'] = pd.cut(df['chol'], bins=bins, labels=labels)

# Count of patients in each category
cholesterol_counts = df['cholesterol_category'].value_counts()

# Plot the results
plt.figure(figsize=(8, 6))
cholesterol_counts.plot(kind='bar', color='teal', edgecolor='black')
plt.title("Number of Patients by Cholesterol Category", fontsize=16)
plt.xlabel("Cholesterol Category", fontsize=12)
plt.ylabel("Number of Patients", fontsize=12)
plt.xticks(rotation=20)
plt.show()
```
